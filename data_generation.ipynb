### combined and short data for all rounds

```python
import pandas as pd

def build_backtest_data(round_num):

    price_data_1 = pd.read_csv(f'data/ROUND_{round_num}/prices_round_{round_num}_day_{round_num-3}.csv', header=0, sep=';')
    price_data_2 = pd.read_csv(f'data/ROUND_{round_num}/prices_round_{round_num}_day_{round_num-2}.csv', header=0, sep=';')
    price_data_3 = pd.read_csv(f'data/ROUND_{round_num}/prices_round_{round_num}_day_{round_num-1}.csv', header=0, sep=';')

    order_data_short = price_data_1[price_data_1['timestamp'] < 200000]
    order_data_short.to_csv(f'data/round{round_num}/prices_short.csv', index=False, sep=';')

    price_data_2['timestamp'] += 1000000
    price_data_3['timestamp'] += 2000000

    order_data = pd.concat([price_data_1, price_data_2, price_data_3])

    trades_data_1 = pd.read_csv(f'data/ROUND_{round_num}/trades_round_{round_num}_day_{round_num-3}.csv', header=0, sep=';')
    trades_data_2 = pd.read_csv(f'data/ROUND_{round_num}/trades_round_{round_num}_day_{round_num-2}.csv', header=0, sep=';')
    trades_data_3 = pd.read_csv(f'data/ROUND_{round_num}/trades_round_{round_num}_day_{round_num-1}.csv', header=0, sep=';')

    trades_data_short = trades_data_1[trades_data_1['timestamp'] < 200000]
    trades_data_short.to_csv(f'data/round{round_num}/trades_short.csv', index=False, sep=';')

    trades_data_2['timestamp'] += 1000000
    trades_data_3['timestamp'] += 2000000

    trades_data = pd.concat([trades_data_1, trades_data_2, trades_data_3])

    order_data.to_csv(f'data/round{round_num}/prices_combined.csv', index=False, sep=';')
    trades_data.to_csv(f'data/round{round_num}/trades_combined.csv', index=False, sep=';')
```

```python
build_backtest_data(3)
```
